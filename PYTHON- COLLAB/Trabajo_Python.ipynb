#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python

def orfs_hebra_positiva(secuencia):
# Con esta función encontraremos todos los ORFS de la hebra positiva
  stop_codones = ["TAA", "TAG", "TGA"] # Codones de Stop
  orfs_positiva = []

  for frame in range(3): # Los codones se leen de 3 en 3 bases
    i = frame # Para probar cada marco de lectura posible
    while i < len(secuencia) - 2: # Evita que se lea un codón incompleto al final
      codon = secuencia[i:i+3] # Buscamos un codón (3 bases)
      if codon == "ATG": # Si el codón es de inicio
        j = i + 3
        encontrado = False
        while j < len(secuencia) -2: # En este bucle buscamos el codón de parada
          stop = secuencia[j:j+3]
          if stop in stop_codones: # Comprobamos que el codón encontrado sea uno de los de stop
            orf = secuencia[i:j+3] # Si lo es, extraemos la secuencia del orf (codón de empiece- codón de stop)
            if orf not in orfs_positiva: # Guardamos el orf siempre que no esté ya en la lista (para evitar duplicados)
              orfs_positiva.append(orf) # Guardamos nuestro orf en nuestra lista
            encontrado = True
            break # Salimos del bucle
          j += 3 # Buscamos el siguiente codón de stop
        i += 3 # Reanudamos la búsqueda de "ATG"
      else:
        i += 3 # En caso de no encontrar el codón de inicio seguimos buscando
  return orfs_positiva

def orfs_hebra_negativa(secuencia):
# Con esta función vamos a buscar los ORFS de la hebra negativa
# Lo primero es obtener la hebra negativa
  complementario = {"A":"T", "T":"A", "G":"C", "C":"G"} # Diccionario con las bases complementarias
  sec_complementaria = "".join(complementario[base] for base in secuencia) # Hacemos la hebra complementaria de la secuencia
  hebra_negativa = sec_complementaria[::-1] # Y le damos la vuelta para sacar la hebra negativa

# Ahora vamos a buscar los ORFS en la hebra negativa reutilizando la función anterior donde hemos sacados los ORFS de la positiva
  orfs_negativa = orfs_hebra_positiva(hebra_negativa)
  return orfs_negativa

def todos_orfs(secuencia):
  orfs_pos = orfs_hebra_positiva(secuencia) # Aquí guardamos los ORFs de la hebra positiva
  orfs_neg = orfs_hebra_negativa(secuencia) # Aquí guardamos los ORFs de la hebra negativa
  # Creamos una lista con todos los ORFS y guardamos de que hebra vienen
  todos_orfs = orfs_pos + orfs_neg
  # Ordenamos nuestra lista de ORFS de mayor a menor según su longitud
  todos_orfs.sort(key = len, reverse = True)
  mejores_6 = todos_orfs[:6] # Seleccionamos los 6 más largos
  mejor_orf = todos_orfs[0] # Seleccionamos el más largo
  return todos_orfs, mejores_6, mejor_orf

def traducir_mejor_orf(mejor_orf):
  # Con esta función buscamos traducir el mejor ORF a proteína
  # Creamos una tabla con codones
  tabla_codones = {
    'ATT':'Ile', 'ATC':'Ile', 'ATA':'Ile', 'ATG':'Met',
    'ACT':'Thr', 'ACC':'Thr', 'ACA':'Thr', 'ACG':'Thr',
    'AAT':'Asn', 'AAC':'Asn', 'AAA':'Lys', 'AAG':'Lys',
    'AGT':'Ser', 'AGC':'Ser', 'AGA':'Arg', 'AGG':'Arg',
    'GTT':'Val', 'GTC':'Val', 'GTA':'Val', 'GTG':'Val',
    'GCT':'Ala', 'GCC':'Ala', 'GCA':'Ala', 'GCG':'Ala',
    'GAT':'Asp', 'GAC':'Asp', 'GAA':'Glu', 'GAG':'Glu',
    'GGT':'Gly', 'GGC':'Gly', 'GGA':'Gly', 'GGG':'Gly',
    'TTT':'Phe', 'TTC':'Phe', 'TTA':'Leu', 'TTG':'Leu',
    'TCT':'Ser', 'TCC':'Ser', 'TCA':'Ser', 'TCG':'Ser',
    'TAT':'Tyr', 'TAC':'Tyr', 'TAA':'Stop','TAG':'Stop',
    'TGT':'Cys', 'TGC':'Cys', 'TGA':'Stop','TGG':'Trp',
    'CTT':'Leu', 'CTC':'Leu', 'CTA':'Leu', 'CTG':'Leu',
    'CCT':'Pro', 'CCC':'Pro', 'CCA':'Pro', 'CCG':'Pro',
    'CAT':'His', 'CAC':'His', 'CAA':'Gln', 'CAG':'Gln',
    'CGT':'Arg', 'CGC':'Arg', 'CGA':'Arg', 'CGG':'Arg',
}
  proteina = [] # Aquí guardaremos los aa
  for i in range(0, len(mejor_orf)-2, 3): # Recorremos la secuencia del ORF de 3 en 3
    codon = mejor_orf[i:i+3] # Tomamos el codón de 3 bases
    aa = tabla_codones.get(codon, 'X') # X si codón no válido
    if aa == "Stop":  # Detenemos la traducción en el codón stop
      break
    proteina.append(aa) # Añadimos el aminoácido a la proteína
  return proteina

def analisis_orf(secuencia):
  # Obtenemos todos los ORFs y el mejor
    todos, mejores_6, mejor = todos_orfs(secuencia)
  # Creamos tuplas con la hebra de cada ORF solo para la impresión
    orfs_pos = orfs_hebra_positiva(secuencia)
    orfs_neg = orfs_hebra_negativa(secuencia)

    todos_orfs_tuplas = [("positiva", orf) if orf in orfs_pos else ("negativa", orf) for orf in todos]
    mejores_6_tuplas = [("positiva", orf) if orf in orfs_pos else ("negativa", orf) for orf in mejores_6]
    mejor_orf_tupla = ("positiva", mejor) if mejor in orfs_pos else ("negativa", mejor)

  # Imprimimos todos los ORFs
    print("Todos los ORFs encontrados:")
    for hebra, orf in todos_orfs_tuplas:
        print(f"Hebra: {hebra} | ORF: {orf} | Longitud: {len(orf)}")

  # Imprimimos los 6 mejores ORFs
    print("\nLos 6 ORFs más largos:")
    for i, (hebra, orf) in enumerate(mejores_6_tuplas, 1):
        print(f"ORF{i}: Hebra: {hebra} | Longitud: {len(orf)} | Secuencia: {orf}")

  # Analizamos el mejor ORF
    hebra_mejor, sec_mejor = mejor_orf_tupla
    proteina = traducir_mejor_orf(sec_mejor)
    aromaticos = ["Phe", "Tyr", "Trp"] # aa aromáticos
  # Primero calculamos la cantidad de aa aromáticos
    cantidad_aromaticos = sum(1 for aa in proteina if aa in aromaticos)
  # Después calculamos el porcentaje dividiendo lo anterior entre la longitud de la proteína
    porcentaje_aromaticos = (cantidad_aromaticos / len(proteina)) * 100
  # Dividimos la longitud de la secuencia del orf entre la longitud de la secuencia original
    longitud_relativa = (len(proteina) / len(secuencia)) * 100

    print("\n-Mejor ORF-")
    print(f"Hebra: {hebra_mejor}")
    print(f"Secuencia ORF: {sec_mejor}")
    print(f"Longitud ORF: {len(sec_mejor)}")
    print(f"Proteína traducida: {'-'.join(proteina)}")
    print(f"Porcentaje aminoácidos aromáticos: {porcentaje_aromaticos} %")
    print(f"Longitud relativa de la secuencia: {longitud_relativa} %")

analisis_orf(secuencia)
```
