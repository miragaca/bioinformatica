#### 5.- De la lista de accessions que se proporciona obtén lo siguiente:
1. Los que tengan el número 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o más dígitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

```python
import re

accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']

print("1. Contenga el número '5':",[t for t in accs if re.search(r'5',t)])
print("2. Contengan 'd' o 'e':",[t for t in accs if re.search(r'[de]',t)])
print("3. Contengan 'd' y 'e' en ese orden:",[t for t in accs if re.search(r'd.*e',t)])
print("4. Contenga 'd' y 'e' en ese orden con cualquier letra en medio:",[t for t in accs if re.search(r'd.e',t)])
print("5. Contengan'd' y 'e' en cualquier orden:",[t for t in accs if re.search(r'd',t) and re.search(r'e',t)])
print("6. Empiezen por 'x' o por 'y':",[t for t in accs if re.search(r'^[xy]',t)])
print("7. Empiezen por 'x' o 'y'y terminen en 'e':",[t for t in accs if re.search(r'^[xy].*e$',t)])
print("8. Tienen tres o mas dígitos seguidos:",[t for t in accs if re.search(r'\d{3,}',t)] )
print("9. Terminen con'da','dr' o 'dp':",[t for t in accs if re.search(r'd[arp]$',t)])
```
